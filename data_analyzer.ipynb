```python
# analyze self rating ratio

mat = scipy.io.loadmat('./ASCERTAIN_Features/Dt_SelfReports.mat')
mat = mat['Ratings']

# arousals
summ = 0
arousals = mat[0]
arousals1 = np.array(arousals)[:,0:3] # 0~2, 5~31 # ! 3, 4, 32
arousals2 = np.array(arousals)[:,5:29]
arousals1 = arousals1.flatten()
print(arousals1)
arousals1 = [1 if x>3 else 0 for x in arousals1]
summ += np.sum(arousals1)
arousals2 = arousals2.flatten()
arousals2 = [1 if x>3 else 0 for x in arousals2]
summ += np.sum(arousals2)
print('HA percentage ratio: ', summ/(len(arousals1) + len(arousals2)))

summ = 0
arousals = mat[0]
arousals1 = np.array(arousals)[:,0:3] # 0~2, 5~31 # ! 3, 4, 32
arousals2 = np.array(arousals)[:,5:29]
arousals1 = arousals1.flatten()
print(arousals1)
arousals1 = [1 if x<3 else 0 for x in arousals1]
summ += np.sum(arousals1)
arousals2 = arousals2.flatten()
arousals2 = [1 if x<3 else 0 for x in arousals2]
summ += np.sum(arousals2)
print('HA percentage ratio: ', summ/(len(arousals1) + len(arousals2)))





clusters = [2, 0, 1, 2, 2, 2, 2, 2, 1, 1, 1, 0, 0, 0, 2, 2, 1, 2, 1, 1, 0, 1, 0, 2, 1, 0, 0, 1, 2, 2, 2, 2, 2, 2, 0, 0, 0,
 1, 2, 2, 0, 2, 1, 2, 0, 0, 1]

arousals = mat[0] # neutral rating
print(arousals.shape)
c_arousals = [[], [], []]
for i in range(47):
    c_arousals[clusters[i]].append(arousals[i])

for arousals in c_arousals:
    arousals = np.array(arousals).flatten()
    arousals = [1 if x<=3 else 0 for x in arousals] 
    print('LA percentage ratio: ', np.sum(arousals)/len(arousals))
arousals = mat[0]

for i in range(34):
    t_arousals = [1 if x > 3 else 0 for x in arousals[:,i]]
    print(str(i) + ' :' + str(np.sum(t_arousals)/58))
    
# arousals
arousals = mat[0]
arousal_means = [np.mean(arousals[:,i]) for i in range(36)]
print(arousal_means)
total = arousals.flatten().shape[0]
positive = 0
for i in range(58):
    for j in range(36):
        if arousals[i][j] >= arousal_means[j]:
            positive += 1
print(positive/total)

arousals = arousals.flatten()
arousals = [1 if x >= 4 else 0 for x in arousals]
print('HA percentage ratio: ', np.sum(arousals)/len(arousals))

arousals = mat[0]
arousals = arousals.flatten()
arousals = [1 if x < 4 else 0 for x in arousals]
print('LA percentage ratio: ', np.sum(arousals)/len(arousals))

print()
# valences
valences = mat[1]
valences = valences.flatten()
valences = [1 if x <= 0 else 0 for x in valences]
print('LV percentage ratio: ', np.sum(valences)/len(valences))

valences = mat[1]
valences = valences.flatten()
valences = [1 if x >= 1 else 0 for x in valences]
print('HV percentage ratio: ', np.sum(valences)/len(valences))

valences = mat[1] # neutral rating
valences = valences.flatten()
valences = [1 if x==0 else 0 for x in valences]
print('NV percentage ratio: ', np.sum(valences)/len(valences))

# Engagement
engagements = mat[2]
engagements = engagements.flatten()
engagements = [1 if x <= 3 else 0 for x in engagements]
print('LE percentage ratio: ', np.sum(engagements)/len(engagements))

# Liking
likings = mat[3]
likings = likings.flatten()
likings = [1 if x <= 3 else 0 for x in likings]
print('LL percentage ratio: ', np.sum(likings)/len(likings))
```

```python
# data analysis for MMAE(Multi-Modality Auto-Encoder)

# intersection of data quality 1~4
f = open('./data_quality/raw_EEG.txt','r')
raw_EEG_quality = 36 * [58 * [0]]
data = f.read()
clips = data.split('\n')
for i in range(36):
    j = 0
    for text in clips[i].split():
        raw_EEG_quality[i][j] = int(text)
        j += 1
        
f = open('./data_quality/ECG.txt','r')
ECG_quality = 36 * [58 * [0]]
data = f.read()
clips = data.split('\n')
for i in range(36):
    j = 0
    for text in clips[i].split():
        ECG_quality[i][j] = int(text)
        j += 1
        
f = open('./data_quality/GSR.txt','r')
GSR_quality = 36 * [58 * [0]]
data = f.read()
clips = data.split('\n')
for i in range(36):
    j = 0
    for text in clips[i].split():
        GSR_quality[i][j] = int(text)
        j += 1

quality_count = 0
for i in range(36):
    for j in range(58):
        if raw_EEG_quality[i][j] <= 4 and ECG_quality[i][j] <= 4 and GSR_quality[i][j] <= 4:
            quality_count += 1
print('qualified data ratio: ', quality_count / (36*58))
```
